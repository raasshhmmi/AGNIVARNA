```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
import nltk
from nltk.tokenize import word_tokenize

# Sample conversation data
data = {
    'conversations': [
        ["I have just taken an ECG reading. What should I look for?", "Look for signs like an irregular heartbeat, skipped beats, or any abnormal patterns such as ST elevation, T-wave inversion, or a prolonged QT interval. If you notice anything unusual, seek medical attention immediately."],
        ["What does an irregular heartbeat on an ECG mean?", "An irregular heartbeat, also known as an arrhythmia, may indicate a variety of conditions, including atrial fibrillation, premature ventricular contractions, or more serious conditions like ventricular tachycardia. If you detect an irregular rhythm, it's important to contact a healthcare provider."],
        ["My ECG shows ST elevation. What does that mean?", "ST elevation can indicate a myocardial infarction (heart attack). This is a medical emergency, and you should call emergency services immediately. While waiting, take aspirin if not allergic and keep calm."]
    ]
}
# Convert list to DataFrame
df = pd.DataFrame(data, columns=['Question', 'Response'])

# Preprocess the data
nltk.download('punkt')
def preprocess(text):
    tokens = word_tokenize(text.lower())
    return ' '.join(tokens)

df['Question'] = df['Question'].apply(preprocess)

# Vectorize the questions
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['Question'])

# Train a Nearest Neighbors model
model = NearestNeighbors(n_neighbors=1, metric='cosine')
model.fit(X)

def chatbot_response(user_input):
    # Preprocess the user input
    user_input_processed = preprocess(user_input)
    user_input_vector = vectorizer.transform([user_input_processed])

    # Find the closest match
    distances, indices = model.kneighbors(user_input_vector)

    # Return the response with the closest match
    response_index = indices[0][0]
    return df['Response'].iloc[response_index]

# Test the chatbot
user_input = "What should I do if my ECG reading shows ST elevation?"
response = chatbot_response(user_input)
print("Chatbot:", response)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
import nltk
from nltk.tokenize import word_tokenize
import gradio as gr

# Sample conversation data
data = {
    'Question': [
        "I have just taken an ECG reading. What should I look for?",
        "What does an irregular heartbeat on an ECG mean?",
        "My ECG shows ST elevation. What does that mean?",
        "What does a prolonged QT interval mean on an ECG?",
        "My ECG shows a flat T-wave. What should I do?",
        "What is a normal heart rate on an ECG?",
        "My heart rate is above 100 bpm. What does it mean?",
        "My heart rate is below 60 bpm. Is this dangerous?",
        "What medication can I take for tachycardia?",
        "What should I do if I feel dizzy and my ECG looks abnormal?",
        "Can anxiety affect my ECG reading?",
        "What is the role of aspirin in case of a suspected heart attack?",
        "Should I take nitroglycerin if I have chest pain?",
        "What should I do if I experience chest pain?",
        "What does atrial fibrillation look like on an ECG?",
        "What medication is recommended for atrial fibrillation?",
        "What should I do if my ECG shows PVCs (Premature Ventricular Contractions)?",
        "What does a heart block look like on an ECG?",
        "What should I do if my ECG shows a heart block?",
        "How can I lower my heart rate naturally?",
        "What is ventricular tachycardia?",
        "What should I do if I have ventricular tachycardia?",
        "Can dehydration affect my ECG reading?",
        "What should I do if I feel short of breath and my ECG is normal?",
        "How do I interpret ST depression on my ECG?",
        "What is Torsades de Pointes?",
        "What should I do if I suspect Torsades de Pointes?",
        "What medications can I take to stabilize my heart rhythm?",
        "How can I prevent arrhythmias?",
        "What are the warning signs of a heart attack?",
        "How should I manage an arrhythmia detected on an ECG?",
        "Can I exercise if I have an arrhythmia?",
        "What is the role of beta-blockers in heart conditions?",
        "What should I do for a first-degree burn?",
        "How can I treat a second-degree burn?",
        "I have just taken an ECG reading. What should I look for?",
        "What does an irregular heartbeat on an ECG mean?",
        "My ECG shows ST elevation. What does that mean?",
        "What does a prolonged QT interval mean on an ECG?",
        "My ECG shows a flat T-wave. What should I do?",
        "What is a normal heart rate on an ECG?",
        "My heart rate is above 100 bpm. What does it mean?",
        "My heart rate is below 60 bpm. Is this dangerous?",
        "What medication can I take for tachycardia?",
        "What should I do if I feel dizzy and my ECG looks abnormal?",
        "Can anxiety affect my ECG reading?",
        "What is the role of aspirin in case of a suspected heart attack?",
        "Should I take nitroglycerin if I have chest pain?",
        "What should I do if I experience chest pain?",
        "What does atrial fibrillation look like on an ECG?",
        "What medication is recommended for atrial fibrillation?",
        "What should I do if my ECG shows PVCs (Premature Ventricular Contractions)?",
        "What does a heart block look like on an ECG?",
        "What should I do if my ECG shows a heart block?",
        "How can I lower my heart rate naturally?",
        "What is ventricular tachycardia?",
        "What should I do if I have ventricular tachycardia?",
        "Can dehydration affect my ECG reading?",
        "What should I do if I feel short of breath and my ECG is normal?",
        "How do I interpret ST depression on my ECG?",
        "What is Torsades de Pointes?",
        "What should I do if I suspect Torsades de Pointes?",
        "What medications can I take to stabilize my heart rhythm?",
        "How can I prevent arrhythmias?",
        "What are the warning signs of a heart attack?",
        "How should I manage an arrhythmia detected on an ECG?",
        "Can I exercise if I have an arrhythmia?",
        "What is the role of beta-blockers in heart conditions?",
        "How can I improve my heart health naturally?",
        "I have just checked my blood pressure. What is considered normal?",
        "My blood pressure is 140/90 mmHg. Is this high?",
        "What should I do if my blood pressure is too high?",
        "What medications can help lower high blood pressure?",
        "My blood pressure is low. What does that mean?",
        "How can I raise my blood pressure if it's too low?",
        "Can stress affect my blood pressure?",
        "What lifestyle changes can help lower blood pressure?",
        "Should I take medication if my blood pressure is mildly elevated?",
        "What are the symptoms of a hypertensive crisis?",
        "Can dehydration cause low blood pressure?",
        "What should I do if my blood pressure fluctuates frequently?",
        "How often should I check my blood pressure at home?",
        "What should I do if I experience dizziness with normal blood pressure?",
        "Can high blood pressure affect my heart rhythm?",
        "What is white-coat hypertension?",
        "How can I reduce white-coat hypertension?",
        "Is it normal for blood pressure to vary throughout the day?",
        "What foods should I avoid to manage my blood pressure?",
        "Can I stop taking my blood pressure medication if my readings are normal?",
        "I have a burn, what should I do?",
        "How do I know if it is a first-degree burn?",
        "How do I know if it is a second-degree burn?",
        "How do I know if it is a third-degree burn?",
        "What should I do for a first-degree burn?",
        "What medication should I use for a first-degree burn?",
        "What should I do for a second-degree burn?",
        "What medication should I use for a second-degree burn?",
        "What should I do for a third-degree burn?",
        "What medication is recommended for a third-degree burn?",
        "What if my burn is infected?",
        "Can I use antibiotic ointment on a burn?",
        "Should I pop a blister caused by a burn?",
        "What pain relief can I take for a burn?",
        "What should I do if my child gets a burn?",
        "What is Silver sulfadiazine cream?",
        "Can I use aloe vera on a burn?",
        "How can I tell if a burn is healing?",
        "How do I prevent scars after a burn?",
        "What should I do for a chemical burn?",
        "What should I do for an electrical burn?",
        "Is there any home remedy for a minor burn?",
        "Can I use honey on a burn?",
        "What should I avoid doing with a burn?",
        "When should I seek medical help for a burn?",
        "How can I reduce pain from a burn?",
        "What should I do for a sunburn?",
        "Can a burn cause dehydration?",
        "Can I use antiseptic cream on a burn?",
        "How do I treat a burn caused by hot water?",
        "Is a burn from steam different from a burn from fire?",
        "Can a burn cause shock?",
        "What should I do if I have a burn on my face?",
        "How do I care for a burn while it is healing?"
    ],
    'Response': [
        "Look for signs like an irregular heartbeat, skipped beats, or any abnormal patterns such as ST elevation, T-wave inversion, or a prolonged QT interval. If you notice anything unusual, seek medical attention immediately.",
        "An irregular heartbeat, also known as an arrhythmia, may indicate a variety of conditions, including atrial fibrillation, premature ventricular contractions, or more serious conditions like ventricular tachycardia. If you detect an irregular rhythm, it's important to contact a healthcare provider.",
        "ST elevation can indicate a myocardial infarction (heart attack). This is a medical emergency, and you should call emergency services immediately. While waiting, take aspirin if not allergic and keep calm.",
        "A prolonged QT interval can increase the risk of a potentially life-threatening arrhythmia called Torsades de Pointes. You should consult a cardiologist for a detailed evaluation and may need to avoid certain medications and triggers.",
        "A flat or inverted T-wave can indicate ischemia, electrolyte imbalance, or other heart conditions. It is important to consult a healthcare provider for further evaluation.",
        "A normal heart rate is typically between 60-100 beats per minute (bpm). However, the normal range can vary based on individual factors such as age, fitness level, and health conditions.",
        "A heart rate above 100 bpm is called tachycardia. It can be caused by stress, dehydration, fever, or more serious heart conditions. Stay calm, rest, and monitor your symptoms. If you experience chest pain, dizziness, or shortness of breath, seek medical attention.",
        "A heart rate below 60 bpm is called bradycardia. It can be normal in athletes or during sleep, but if accompanied by dizziness, fainting, or weakness, seek medical advice.",
        "If tachycardia is due to a non-serious cause like anxiety or dehydration, try relaxation techniques and rehydrate. For persistent tachycardia, beta-blockers like metoprolol or calcium channel blockers like verapamil may be prescribed, but consult a healthcare provider first.",
        "Sit or lie down immediately, and drink some water. Avoid driving or operating machinery. If the dizziness persists, is accompanied by chest pain, shortness of breath, or any other concerning symptoms, seek medical attention immediately.",
        "Yes, anxiety can cause rapid heart rate and palpitations, which may be reflected on your ECG as tachycardia or premature beats. Practice deep breathing and relaxation techniques, and consult a healthcare provider if the symptoms persist.",
        "Aspirin can help thin the blood and prevent further clotting in the case of a suspected heart attack. If you are not allergic to aspirin, take a dose of 325 mg while waiting for emergency medical services.",
        "If you have been prescribed nitroglycerin for chest pain, take it as directed. If not prescribed, do not take it without consulting a doctor, as it can cause significant drops in blood pressure.",
        "Stop any physical activity, sit or lie down, and try to stay calm. Take aspirin if not allergic and contact emergency services immediately. Do not drive yourself to the hospital.",
        "Atrial fibrillation is characterized by irregular and often rapid heart rhythm with no distinct P waves and irregular R-R intervals. It increases the risk of stroke, and you should seek medical advice if detected.",
        "Common medications include anticoagulants (e.g., warfarin, rivaroxaban) to reduce stroke risk, beta-blockers (e.g., metoprolol) or calcium channel blockers (e.g., diltiazem) for rate control, and antiarrhythmic drugs (e.g., amiodarone) to restore normal rhythm.",
        "PVCs are common and usually benign if occasional. Reduce caffeine and alcohol intake, manage stress, and monitor your symptoms. If you experience frequent PVCs or they are associated with other symptoms, consult a healthcare provider.",
        "Heart block can appear as prolonged PR intervals, dropped beats, or complete dissociation between P waves and QRS complexes. A healthcare provider should evaluate any signs of heart block.",
        "First-degree heart block might not need treatment, but second or third-degree heart block can be serious. Avoid certain medications and activities, and consult a cardiologist for further evaluation.",
        "To lower your heart rate naturally, try deep breathing exercises, relaxation techniques, staying hydrated, avoiding stimulants like caffeine, and engaging in regular exercise. If your heart rate is consistently high, consult a doctor.",
        "Ventricular tachycardia is a fast heart rhythm that starts in the heart's ventricles. It is potentially life-threatening and requires immediate medical attention.",
        "Ventricular tachycardia is a medical emergency. Call emergency services immediately. Stay calm, lie down, and wait for help to arrive.",
        "Yes, dehydration can cause electrolyte imbalances, leading to abnormal ECG readings such as changes in the T-wave or QT interval. Drink water and seek medical advice if symptoms persist.",
        "Shortness of breath can be caused by many factors, including anxiety, respiratory conditions, or other non-cardiac causes. Try to stay calm, breathe slowly, and seek medical advice if symptoms worsen or persist.",
        "ST depression can indicate ischemia (reduced blood flow to the heart). This should be evaluated by a healthcare provider, as it may require further testing or treatment.",
        "Torsades de Pointes is a specific type of ventricular tachycardia characterized by rapid, irregular QRS complexes. It is life-threatening and requires immediate medical intervention.",
        "Seek emergency medical help immediately. While waiting, avoid any triggers like certain medications, and keep calm.",
        "Medications to stabilize heart rhythm include beta-blockers, calcium channel blockers, and antiarrhythmic drugs like amiodarone or sotalol. Always consult a healthcare provider before taking any new medication.",
        "To prevent arrhythmias, maintain a healthy lifestyle by managing stress, avoiding stimulants like caffeine, alcohol, and tobacco, eating a heart-healthy diet, and exercising regularly. If prescribed, take medications as directed by your doctor.",
        "Warning signs of a heart attack include chest pain or discomfort, pain in the arms, back, neck, jaw, or stomach, shortness of breath, cold sweat, nausea, and lightheadedness. Seek emergency medical attention immediately if you experience these symptoms.",
        "If you detect an arrhythmia, try to stay calm, sit or lie down, and monitor your symptoms. Avoid stimulants, and seek medical advice. For severe symptoms like fainting or chest pain, call emergency services immediately.",
        "Light to moderate exercise is often beneficial, but it depends on the type and severity of the arrhythmia. Consult your healthcare provider to determine a safe level of activity.",
        "Beta-blockers reduce heart rate and blood pressure, improving heart function and preventing arrhythmias. Common beta-blockers include metoprolol, atenolol, and propranolol. They should only be taken under medical supervision.",
        "You should cool the burn under running cold water for 10-15 minutes.",
        "You should seek medical attention. Second-degree burns require more advanced treatment.",
        "Look for signs like an irregular heartbeat, skipped beats, or any abnormal patterns such as ST elevation, T-wave inversion, or a prolonged QT interval. If you notice anything unusual, seek medical attention immediately.",
        "An irregular heartbeat, also known as an arrhythmia, may indicate a variety of conditions, including atrial fibrillation, premature ventricular contractions, or more serious conditions like ventricular tachycardia. If you detect an irregular rhythm, it's important to contact a healthcare provider.",
        "ST elevation can indicate a myocardial infarction (heart attack). This is a medical emergency, and you should call emergency services immediately. While waiting, take aspirin if not allergic and keep calm.",
        "A prolonged QT interval can increase the risk of a potentially life-threatening arrhythmia called Torsades de Pointes. You should consult a cardiologist for a detailed evaluation and may need to avoid certain medications and triggers.",
        "A flat or inverted T-wave can indicate ischemia, electrolyte imbalance, or other heart conditions. It is important to consult a healthcare provider for further evaluation.",
        "A normal heart rate is typically between 60-100 beats per minute (bpm). However, the normal range can vary based on individual factors such as age, fitness level, and health conditions.",
        "A heart rate above 100 bpm is called tachycardia. It can be caused by stress, dehydration, fever, or more serious heart conditions. Stay calm, rest, and monitor your symptoms. If you experience chest pain, dizziness, or shortness of breath, seek medical attention.",
        "A heart rate below 60 bpm is called bradycardia. It can be normal in athletes or during sleep, but if accompanied by dizziness, fainting, or weakness, seek medical advice.",
        "If tachycardia is due to a non-serious cause like anxiety or dehydration, try relaxation techniques and rehydrate. For persistent tachycardia, beta-blockers like metoprolol or calcium channel blockers like verapamil may be prescribed, but consult a healthcare provider first.",
        "Sit or lie down immediately, and drink some water. Avoid driving or operating machinery. If the dizziness persists, is accompanied by chest pain, shortness of breath, or any other concerning symptoms, seek medical attention immediately.",
        "Yes, anxiety can cause rapid heart rate and palpitations, which may be reflected on your ECG as tachycardia or premature beats. Practice deep breathing and relaxation techniques, and consult a healthcare provider if the symptoms persist.",
        "Aspirin can help thin the blood and prevent further clotting in the case of a suspected heart attack. If you are not allergic to aspirin, take a dose of 325 mg while waiting for emergency medical services.",
        "If you have been prescribed nitroglycerin for chest pain, take it as directed. If not prescribed, do not take it without consulting a doctor, as it can cause significant drops in blood pressure.",
        "Stop any physical activity, sit or lie down, and try to stay calm. Take aspirin if not allergic and contact emergency services immediately. Do not drive yourself to the hospital.",
        "Atrial fibrillation is characterized by irregular and often rapid heart rhythm with no distinct P waves and irregular R-R intervals. It increases the risk of stroke, and you should seek medical advice if detected.",
        "Common medications include anticoagulants (e.g., warfarin, rivaroxaban) to reduce stroke risk, beta-blockers (e.g., metoprolol) or calcium channel blockers (e.g., diltiazem) for rate control, and antiarrhythmic drugs (e.g., amiodarone) to restore normal rhythm.",
        "PVCs are common and usually benign if occasional. Reduce caffeine and alcohol intake, manage stress, and monitor your symptoms. If you experience frequent PVCs or they are associated with other symptoms, consult a healthcare provider.",
        "Heart block can appear as prolonged PR intervals, dropped beats, or complete dissociation between P waves and QRS complexes. A healthcare provider should evaluate any signs of heart block.",
        "First-degree heart block might not need treatment, but second or third-degree heart block can be serious. Avoid certain medications and activities, and consult a cardiologist for further evaluation.",
        "To lower your heart rate naturally, try deep breathing exercises, relaxation techniques, staying hydrated, avoiding stimulants like caffeine, and engaging in regular exercise. If your heart rate is consistently high, consult a doctor.",
        "Ventricular tachycardia is a fast heart rhythm that starts in the heart's ventricles. It is potentially life-threatening and requires immediate medical attention.",
        "Ventricular tachycardia is a medical emergency. Call emergency services immediately. Stay calm, lie down, and wait for help to arrive.",
        "Yes, dehydration can cause electrolyte imbalances, leading to abnormal ECG readings such as changes in the T-wave or QT interval. Drink water and seek medical advice if symptoms persist.",
        "Shortness of breath can be caused by many factors, including anxiety, respiratory conditions, or other non-cardiac causes. Try to stay calm, breathe slowly, and seek medical advice if symptoms worsen or persist.",
        "ST depression can indicate ischemia (reduced blood flow to the heart). This should be evaluated by a healthcare provider, as it may require further testing or treatment.",
        "Torsades de Pointes is a specific type of ventricular tachycardia characterized by rapid, irregular QRS complexes. It is life-threatening and requires immediate medical intervention.",
        "Seek emergency medical help immediately. While waiting, avoid any triggers like certain medications, and keep calm.",
        "Medications to stabilize heart rhythm include beta-blockers, calcium channel blockers, and antiarrhythmic drugs like amiodarone or sotalol. Always consult a healthcare provider before taking any new medication.",
        "To prevent arrhythmias, maintain a healthy lifestyle by managing stress, avoiding stimulants like caffeine, alcohol, and tobacco, eating a heart-healthy diet, and exercising regularly. If prescribed, take medications as directed by your doctor.",
        "Warning signs of a heart attack include chest pain or discomfort, pain in the arms, back, neck, jaw, or stomach, shortness of breath, cold sweat, nausea, and lightheadedness. Seek emergency medical attention immediately if you experience these symptoms.",
        "If you detect an arrhythmia, try to stay calm, sit or lie down, and monitor your symptoms. Avoid stimulants, and seek medical advice. For severe symptoms like fainting or chest pain, call emergency services immediately.",
        "Light to moderate exercise is often beneficial, but it depends on the type and severity of the arrhythmia. Consult your healthcare provider to determine a safe level of activity.",
        "Beta-blockers reduce heart rate and blood pressure, improving heart function and preventing arrhythmias. Common beta-blockers include metoprolol, atenolol, and propranolol. They should only be taken under medical supervision.",
        "To improve heart health, eat a balanced diet, exercise regularly, avoid smoking, limit alcohol, and manage stress. Maintaining a healthy weight and keeping your blood pressure, cholesterol, and diabetes under control are also important.",
        "Normal blood pressure is typically around 120/80 mmHg. However, individual targets may vary based on age, health conditions, and doctor's advice.",
        "Yes, a reading of 140/90 mmHg or higher is considered hypertension (high blood pressure). You should monitor your readings regularly and consult a healthcare provider for advice on management.",
        "If your blood pressure is high, sit down and relax. Avoid stress, caffeine, and salt. Consider taking medication if prescribed. If your blood pressure is severely elevated (e.g., above 180/120 mmHg) and you experience symptoms like chest pain, call emergency services immediately.",
        "Common medications include ACE inhibitors (e.g., lisinopril), ARBs (e.g., losartan), calcium channel blockers (e.g., amlodipine), diuretics (e.g., hydrochlorothiazide), and beta-blockers (e.g., metoprolol). Your doctor will determine the most suitable medication based on your condition.",
        "Low blood pressure (hypotension) is usually defined as a reading below 90/60 mmHg. It may not be concerning if you have no symptoms, but if you feel dizzy, lightheaded, or faint, consult a healthcare provider.",
        "To raise low blood pressure, drink more water, eat small, frequent meals, increase salt intake (if advised by a doctor), avoid standing for long periods, and wear compression stockings if needed. If symptoms persist, seek medical advice.",
        "Yes, stress can cause temporary spikes in blood pressure. Practice relaxation techniques like deep breathing, meditation, and exercise to help manage stress and keep your blood pressure under control.",
        "Lifestyle changes to lower blood pressure include eating a balanced diet low in salt, exercising regularly, maintaining a healthy weight, reducing alcohol intake, quitting smoking, managing stress, and getting enough sleep.",
        "For mildly elevated blood pressure, lifestyle changes like diet and exercise may be recommended first. Medications may be necessary if lifestyle changes are insufficient or if you have other risk factors. Consult your doctor for personalized advice.",
        "A hypertensive crisis is a severe increase in blood pressure (above 180/120 mmHg) that can cause headaches, shortness of breath, nosebleeds, and anxiety. If you experience these symptoms, seek emergency medical help.",
        "Yes, dehydration can lead to low blood pressure. Drink plenty of fluids, especially in hot weather or after exercise, to maintain proper hydration levels.",
        "Monitor your blood pressure at different times of the day and keep a log of your readings. Try to identify any triggers like stress, caffeine, or certain medications. Consult a healthcare provider for further evaluation and guidance.",
        "If you have been diagnosed with hypertension or are at risk, check your blood pressure daily or as recommended by your doctor. For general monitoring, checking a few times a week may be sufficient.",
        "Dizziness can have various causes, such as dehydration, anxiety, or inner ear issues, even with normal blood pressure. Sit or lie down until it passes, and consult a healthcare provider if it persists.",
        "Yes, high blood pressure can contribute to the development of arrhythmias, such as atrial fibrillation. Managing blood pressure is important to reduce the risk of rhythm disturbances.",
        "White-coat hypertension occurs when blood pressure readings are elevated in a medical setting due to anxiety but normal at home. Consider home monitoring to confirm your average blood pressure.",
        "Practice relaxation techniques like deep breathing or meditation before taking a reading. Monitoring your blood pressure at home can also provide a more accurate reflection of your usual levels.",
        "Yes, blood pressure naturally fluctuates due to factors like activity level, stress, and diet. It's usually lower in the morning and can increase with stress or exercise. Consistently high readings should be discussed with your doctor.",
        "To manage blood pressure, avoid foods high in salt, trans fats, and cholesterol, such as processed foods, fast food, and salty snacks. Limit alcohol, and opt for a balanced diet rich in fruits, vegetables, whole grains, and lean proteins.",
        "Do not stop taking your medication without consulting your doctor, even if your blood pressure is normal. Medications help maintain control, and stopping them abruptly can lead to a dangerous spike.",
        "First, determine the severity of the burn. Is it a first-degree, second-degree, or third-degree burn? You can describe the symptoms if you're unsure.",
        "A first-degree burn affects only the outer layer of skin. It causes redness, minor swelling, and pain but does not blister.",
        "A second-degree burn affects both the outer layer and the underlying layer of skin, causing pain, redness, swelling, and blistering.",
        "A third-degree burn affects all layers of the skin and may extend into deeper tissues. The skin may look white, charred, or leathery, and there might be little or no pain due to nerve damage.",
        "For a first-degree burn, cool the area with cool (not cold) water for 10-15 minutes, apply aloe vera or a moisturizing lotion, and take over-the-counter pain relief like ibuprofen or acetaminophen.",
        "You can use topical treatments like aloe vera gel, antibiotic ointment, or a moisturizing lotion. For pain relief, use over-the-counter painkillers like ibuprofen or acetaminophen.",
        "Do not break any blisters. Cool the burn with cool water for 10-15 minutes, cover it with a clean, non-stick bandage, and seek medical advice. Consider using a burn cream like Silver sulfadiazine, and take pain relief like ibuprofen or acetaminophen.",
        "Topical burn ointments like Silver sulfadiazine cream or Bacitracin can help prevent infection. Oral pain relievers like ibuprofen or acetaminophen can help manage pain. Avoid using ice, butter, or oil.",
        "Third-degree burns require immediate medical attention. Call emergency services or visit the nearest hospital. Do not apply water or any ointment, and cover the burn with a sterile, non-stick bandage or clean cloth.",
        "Third-degree burns need professional medical treatment, which may include intravenous fluids, antibiotics to prevent infection, and pain management medications like opioids.",
        "Signs of infection include increased pain, redness, swelling, pus, and fever. You should see a doctor immediately. They may prescribe antibiotics, either oral or topical.",
        "Yes, for minor burns, applying an over-the-counter antibiotic ointment like Bacitracin or Neosporin can help prevent infection.",
        "No, do not pop a blister. Blisters protect the underlying skin and help with healing. If a blister breaks, clean it gently with water, apply an antibiotic ointment, and cover it with a clean bandage.",
        "For minor burns, over-the-counter pain relievers like ibuprofen (Advil) or acetaminophen (Tylenol) can help manage pain.",
        "Immediately cool the burn with lukewarm water for 10-15 minutes, avoid using ice, and seek medical attention, especially if the burn is larger than the child's palm or on a sensitive area like the face, hands, or genitals.",
        "Silver sulfadiazine cream is a topical antibiotic used to prevent and treat infections in second and third-degree burns. It helps to stop the growth of bacteria on the wound.",
        "Yes, aloe vera is effective for soothing first-degree burns. It can help reduce pain, redness, and inflammation. Make sure to use pure aloe vera gel.",
        "A healing burn will show less redness and swelling, start to peel as new skin forms, and become less painful over time. If there are signs of infection or worsening, seek medical attention.",
        "To prevent scars, keep the wound clean, moisturize regularly, avoid direct sun exposure, and use silicone gel sheets or scar creams after the wound has healed.",
        "Immediately flush the affected area with large amounts of water for at least 15-20 minutes, remove contaminated clothing, and seek medical help. Do not apply creams or ointments until advised by a healthcare professional.",
        "Electrical burns require immediate medical attention. Call emergency services and do not touch the person if they are still in contact with the electrical source. Turn off the power source first.",
        "For first-degree burns, you can use cool water, aloe vera gel, honey, and loose gauze to cover the wound. Avoid using ice, butter, or greasy ointments.",
        "Yes, honey has natural antibacterial properties and can be used for minor burns to prevent infection and soothe the wound.",
        "Avoid using ice, butter, oil, or greasy ointments on burns. Do not pop blisters or remove burnt clothing stuck to the skin.",
        "Seek medical help if the burn is larger than 3 inches, is on the face, hands, feet, or genitals, or if it looks infected, is a third-degree burn, or is caused by chemicals or electricity.",
        "You can reduce pain by applying cool compresses, taking over-the-counter pain relievers like ibuprofen or acetaminophen, and using topical treatments like aloe vera or antibiotic ointment for minor burns.",
        "For a sunburn, cool the skin with cool water, apply aloe vera or moisturizing lotion, take pain relievers like ibuprofen, and drink plenty of fluids to stay hydrated.",
        "Yes, severe burns can cause fluid loss and dehydration. It's important to drink plenty of fluids, especially water, to stay hydrated.",
        "For minor burns, antiseptic creams like Neosporin can help prevent infection. However, for more severe burns, consult a healthcare provider before applying any creams.",
        "Immediately cool the area with cool water for 10-15 minutes, cover it with a clean, non-stick bandage, and take pain relievers like ibuprofen. Seek medical advice if the burn is severe.",
        "Yes, a steam burn is typically a scald, caused by hot steam, while a fire burn is caused by direct contact with flames. Both can vary in severity and may need different treatments.",
        "Yes, severe burns, especially third-degree burns, can cause shock due to fluid loss and pain. Signs of shock include pale skin, rapid breathing, and confusion. Seek emergency help immediately.",
        "Burns on the face should always be evaluated by a healthcare professional due to the risk of airway involvement and scarring. Apply a clean, moist cloth and seek medical attention immediately.",
        "Keep the burn clean and covered, apply prescribed ointments or creams, avoid sun exposure, and follow up with your healthcare provider. Change bandages as directed and watch for signs of infection."

    ]
}


# Convert list to DataFrame
df = pd.DataFrame(data, columns=['Question', 'Response'])

# Preprocess the data
nltk.download('punkt')
def preprocess(text):
    tokens = word_tokenize(text.lower())
    return ' '.join(tokens)

df['Question'] = df['Question'].apply(preprocess)

# Vectorize the questions
vectorizer = TfidfVectorizer(stop_words=None)
X = vectorizer.fit_transform(df['Question'])

# Train a Nearest Neighbors model
model = NearestNeighbors(n_neighbors=1, metric='cosine')
model.fit(X)

def chatbot_response(user_input):
    # Preprocess the user input
    user_input_processed = preprocess(user_input)
    user_input_vector = vectorizer.transform([user_input_processed])

    # Find the closest match
    distances, indices = model.kneighbors(user_input_vector)

    # Return the response with the closest match
    response_index = indices[0][0]
    return df['Response'].iloc[response_index]

# Define Gradio interface
def gradio_chatbot(user_input):
    return chatbot_response(user_input)

iface = gr.Interface(fn=gradio_chatbot,
                     inputs="text",
                     outputs="text",
                     title="Medical Chatbot",
                     description="Ask questions about ECG readings and get responses.")

# Launch the interface
iface.launch()
```
